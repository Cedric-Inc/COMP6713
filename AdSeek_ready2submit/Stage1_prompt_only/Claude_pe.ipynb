```python
!pip install anthropic
!pip install tqdm
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import time
import anthropic
from tqdm import tqdm
import json

client = anthropic.Anthropic(
    api_key=""
)

MODEL = "claude-3-5-sonnet-20241022"

def claude_generate(prompt, max_tokens=128, retries=3, delay=2):
    for _ in range(retries):
        try:
            start = time.time()
            response = client.messages.create(
                model=MODEL,
                max_tokens=max_tokens,
                temperature=0.0,
                messages=[{"role": "user", "content": prompt}]
            )
            elapsed = time.time() - start
            return response.content[0].text.strip(), elapsed
        except Exception as e:
            print(f"Error: {e} — Retrying in {delay}s...")
            time.sleep(delay)
    return "CLAUDE_API_ERROR", 0.0

df = pd.read_json("/content/drive/MyDrive/AdSeek/Preprocessing/p_engineering_testsets/p_engineering_testset.json")
# df = pd.concat([df.iloc[:10], df.iloc[600:610], df.iloc[-10:]], ignore_index=True)

responses = []
times = []

for prompt in tqdm(df['prompt']):
    output, t = claude_generate(prompt)
    responses.append(output)
    times.append(t)

df['y_pred'] = responses
df['inference_time'] = times

df.to_json("claude35_predictions_pe.json", orient="records", force_ascii=False, indent=2)
print("✅ Claude 推理结果已保存！")
```
